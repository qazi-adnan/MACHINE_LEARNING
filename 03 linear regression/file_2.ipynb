```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv(r"D:\STUDY\MACHINE_LEARNING\3 linear regression\random_values_y.csv")
```

```python
df
```

```python
df.dtypes
```

```python
x=df["x1"]
y=df["y"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
x_train = pd.DataFrame(x_train)  # Convert Series to DataFrame
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
x_test = pd.DataFrame(x_test)  # Convert Series to DataFrame
```

```python
y_pred=regression.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score


# Calculate MAE
mae = mean_absolute_error(y_test, y_pred)

# Calculate MSE
mse = mean_squared_error(y_test, y_pred)

# Calculate RMSE
rmse = np.sqrt(mse)

score=r2_score(y_test,y_pred)
# Print the results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R_squar score is :{score}")
```
