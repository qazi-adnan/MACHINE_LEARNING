# MULTI CLASS CLASSIFICATION

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\STUDY\MACHINE_LEARNING\5 logsitic regression\IRIS.csv")
```

```python
df
```

```python
df["species"].unique()
```

```python
df["species"]=df["species"].map({'Iris-setosa':1 , 'Iris-versicolor':2, 'Iris-virginica':0})
```

```python
df
```

```python
df["species"].value_counts()
```

```python
x=df.drop("species", axis=1)
y=df["species"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test =train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train_scaled,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(model , filled=True)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
score=accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix, classification_report 

model=DecisionTreeClassifier(max_depth=2)
model.fit(x_train_scaled,y_train)


from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(model , filled=True)
y_pred=model.predict(x_test_scaled)


score=accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python

from sklearn.metrics import accuracy_score,confusion_matrix, classification_report 

model=DecisionTreeClassifier(max_depth=3)
model.fit(x_train_scaled,y_train)


from sklearn import tree
plt.figure(figsize=(10,8))
tree.plot_tree(model , filled=True)
y_pred=model.predict(x_test_scaled)


score=accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

# DECESION TREE CLASSFIER PREPRUNING AND HYPER PARAMETER TUNING

```python
param={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel=DecisionTreeClassifier()
```

```python
grid=GridSearchCV(treemodel,param_grid=param,cv=5,scoring='accuracy')
```

```python
grid.fit(x_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
```

```python
score=accuracy_score(y_test, y_pred)
print(score)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```
