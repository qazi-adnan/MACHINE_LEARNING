```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
df["sex"]=le1.fit_transform(df["sex"])
df["smoker"]=le2.fit_transform(df["smoker"])
df["time"]=le3.fit_transform(df["time"])
```

```python
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer(transformers=[('onehot',
                                    OneHotEncoder(drop='first'),
                                    [3])],
                                   remainder='passthrough')
```

```python
df
```

```python
x=df.drop("total_bill", axis=1)
```

```python
y=df["total_bill"]
```

```python
print(type(x))
```

```python
x
```

```python
x=x.to_numpy()
```

```python
x
```

```python
x=ct.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size=0.2 , random_state=42)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)

```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print("r2 score is ", r2_score(y_test,y_pred))
print("mae is" ,mean_absolute_error(y_test,y_pred))
```

# HYPER PARAMETER TUNING 

```python
from sklearn.model_selection import GridSearchCV
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction=grid.predict(x_test)
```

```python
print(r2_score(y_test,grid_prediction))
print(mean_absolute_error(y_test,grid_prediction))
```
